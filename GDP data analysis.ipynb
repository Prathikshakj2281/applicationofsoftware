## Categorise the states into four groups based on the GDP per capita (C1, C2, C3, C4)

```python
# Extract GDP per capita data
gdp_per_capita = filtered_df[filtered_df['Item'] == 'Per Capita GSDP (Rs.)']

# Sort by GDP per capita for clarity
gdp_per_capita = gdp_per_capita.sort_values(by='2014-15', ascending=False)

# Define quantile values
quantiles = gdp_per_capita['2014-15'].quantile([0.20, 0.50, 0.85, 1.0])

# Define a function to categorize based on quantiles
def categorize_gdp(value):
    if value > quantiles[0.85]:
        return 'C1'
    elif value > quantiles[0.50]:
        return 'C2'
    elif value > quantiles[0.20]:
        return 'C3'
    else:
        return 'C4'

# Apply categorization
gdp_per_capita['Category'] = gdp_per_capita['2014-15'].apply(categorize_gdp)

# Display the results
print(gdp_per_capita[['State', '2014-15', 'Category']])

# Optionally, save the results to a file
gdp_per_capita[['State', '2014-15', 'Category']].to_csv('gdp_per_capita_categories.csv', index=False)
```

## The top sub-sectors (such as agriculture, forestry and fishing, crops, manufacturing etc.) that contribute to approximately 80% of the GSDP of each category.

```python
import pandas as pd

# Define the relevant sub-sectors
sub_sectors = [
    'Agriculture, forestry and fishing',
    'Mining and quarrying',
    'Manufacturing',
    'Electricity, gas, water supply & other utility services',
    'Construction',
    'Trade, repair, hotels and restaurants',
    'Transport, storage, communication & services related to broadcasting',
    'Financial services',
    'Real estate, ownership of dwelling & professional services',
    'Public administration',
    'Other services'
]

# Filter the dataset for relevant sub-sectors
filtered_data = filtered_df[filtered_df['Item'].isin(sub_sectors)]

# Merge the category data with the filtered data
merged_data = pd.merge(filtered_data, gdp_per_capita[['State', 'Category']], on='State')

# Initialize an empty list for the final result
expanded_result = []

# Group by category and state
for (category, state), group in merged_data.groupby(['Category', 'State']):
    # Calculate the total GSDP for the state
    total_gsdp = group['2014-15'].sum()
    
    # Calculate the contribution of each sub-sector
    group['Contribution'] = group['2014-15'] / total_gsdp
    
    # Sort sub-sectors by contribution and calculate cumulative contribution
    group = group.sort_values(by='Contribution', ascending=False)
    group['Cumulative Contribution'] = group['Contribution'].cumsum()
    
    # Identify top sub-sectors contributing approximately 80%
    top_sub_sectors = group[group['Cumulative Contribution'] <= 0.8]
    if top_sub_sectors['Cumulative Contribution'].sum() < 0.8 and len(group) > len(top_sub_sectors):
        top_sub_sectors = pd.concat([top_sub_sectors, group.iloc[[len(top_sub_sectors)]]])
    
    # Add each sub-sector as a separate row
    cumulative_value = 0
    cumulative_percentage = 0
    for _, row in top_sub_sectors.iterrows():
        cumulative_value += row['2014-15']
        cumulative_percentage += row['Contribution']
        expanded_result.append({
            'Category': category,
            'State': state,
            'Sub-Sector': row['Item'],
            'Contribution to GSDP': row['2014-15'],
            'Cumulative Value': cumulative_value,
            'Cumulative Percentage': round(cumulative_percentage * 100, 2)
        })

# Convert the expanded results into a DataFrame
expanded_df = pd.DataFrame(expanded_result)

# Sort the output by category, state, and cumulative percentage
expanded_df = expanded_df.sort_values(by=['Category', 'State', 'Cumulative Percentage'])

# Save or display the result
expanded_df.to_csv('expanded_top_sub_sectors.csv', index=False)
expanded_df
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your data (replace these with your actual file paths if needed)
# gdp_per_capita_sorted = pd.read_csv('gdp_per_capita_data.csv')  # Assuming this is your GDP data
# edu_df = pd.read_csv('education_data.csv')  # Assuming this is your Education data

# Sample structure to align data as per the above
# gdp_per_capita_sorted contains columns like ['S.No.', 'Item', '2014-15', 'State']
# edu_df contains columns like ['Sl. No.', 'Level of Education - State', 'Primary - 2014-2015', etc.]

# Step 1: Rename the necessary columns for easier reference
gdp_per_capita_sorted = gdp_per_capita_sorted.rename(columns={'2014-15': 'GDP per capita', 'State': 'State'})

# Step 2: Select the columns for merging
gdp_per_capita_sorted = gdp_per_capita_sorted[['State', 'GDP per capita']]

# Step 3: Select relevant columns from the education dataset for 2014-2015 dropout rates
edu_df = edu_df.rename(columns={'Level of Education - State': 'State'})
education_cols = ['Primary - 2014-2015', 'Upper Primary - 2014-2015', 'Secondary - 2014-2015', 'Senior Secondary - 2014-2015']

# Step 4: Merge the datasets on 'State'
merged_df = pd.merge(gdp_per_capita_sorted, edu_df[['State'] + education_cols], on='State')

# Step 5: Clean the data by removing rows with missing values (drop NaN values)
merged_df_clean = merged_df.dropna(subset=['GDP per capita'] + education_cols)

# Step 6: Calculate the correlation matrix between GDP per capita and the dropout rates
correlation = merged_df_clean[['GDP per capita'] + education_cols].corr()

# Step 7: Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix between GDP per capita and Dropout Rates (2014-2015)')
plt.show()

# Step 8: Scatter plots for each education level dropout rate vs GDP per capita
education_levels = ['Primary - 2014-2015', 'Upper Primary - 2014-2015', 'Secondary - 2014-2015', 'Senior Secondary - 2014-2015']

# Create scatter plots for each education level
for level in education_levels:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(x=merged_df_clean['GDP per capita'], y=merged_df_clean[level])
    plt.title(f'GDP per capita vs {level} Dropout Rates (2014-2015)')
    plt.xlabel('GDP per capita (Rs.)')
    plt.ylabel(f'{level} Dropout Rate (%)')
    plt.show()
```

### Higher Dropout Rates in Lower-GDP States: States with lower GDP tend to have higher dropout rates due to financial constraints and limited educational resources.

### Regional Variation: The relationship between GDP and dropout rates differs across states, indicating other factors like cultural influences, education policies, and infrastructure also play a role.

### Primary Education: Stronger correlation between GDP and dropout rates at the primary level.
